```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'

#数据的列名
columns = data.columns

#属性名的前两个字母，代表国家
country_list=[x[:2] for x in columns ][2:]

#国家含有的指标个数计数
a={}
for i in country_list:
    a[i] = country_list.count(i)

#国家名称集合
country_name=list(a.keys())
#国家指标个数集合
country_index = list(a.values())

#国家索引
index_columns = 1
index_list=0
for i in country_index:
    index_columns+=i
    country_index[index_list] = range(index_columns-i+1,index_columns+1)
    index_list+=1
country_index[0] = range(1,6)
country_index
    
country_dict=dict(zip(list(a.keys()),country_index))
```
